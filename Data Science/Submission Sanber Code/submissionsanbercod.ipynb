```
# data analysis
import pandas as pd
import numpy as np
import random as rnd
import statistics

# visualization
import seaborn as sns
sns.set(style='white', context='notebook', palette='deep')
import matplotlib.pyplot as plt
%matplotlib inline


# machine learning
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, ExtraTreesClassifier, VotingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score, StratifiedKFold, learning_curve


```

```
data_train = pd.read_csv('/content/train.csv', index_col='id')
data_test = pd.read_csv('/content/test.csv', index_col='id')

data_train
```

```
data_train.describe()
```

```
data_train.isnull().sum()
```

```
columnsForDrop = ['Pendidikan','Status Perkawinan','Pekerjaan','Keuntungan Kapital','Kerugian Capital']
data_train.drop(columns=columnsForDrop, inplace=True)
################################
data_test.drop(columns=columnsForDrop, inplace=True)
data_train
```

```
data_train.rename({'Jenis Kelamin': 'JK'}, axis=1, inplace=True)
data_test.rename({'Jenis Kelamin': 'JK'}, axis=1, inplace=True)
data_train.rename({'Jam per Minggu': 'JPM'}, axis=1, inplace=True)
data_test.rename({'Jam per Minggu': 'JPM'}, axis=1, inplace=True)
data_train.rename({'Berat Akhir': 'BA'}, axis=1, inplace=True)
data_test.rename({'Berat Akhir': 'BA'}, axis=1, inplace=True)
data_train.rename({'Kelas Pekerja': 'KP'}, axis=1, inplace=True)
data_test.rename({'Kelas Pekerja': 'KP'}, axis=1, inplace=True)

data_train
```

```
data_train.JK.replace('Laki2', 0, inplace=True)
data_train.JK.replace('Perempuan', 1, inplace=True)
data_train.KP.replace('?', 1, inplace=True)
data_train.KP.replace('Wiraswasta', 1, inplace=True)
data_train.KP.replace('Pekerja Bebas Perusahaan', 2, inplace=True)
data_train.KP.replace('Pekerja Bebas Bukan Perusahan', 2, inplace=True)
data_train.KP.replace('Pemerintah Lokal', 3, inplace=True)
data_train.KP.replace('Pemerintah Provinsi', 3, inplace=True)
data_train.KP.replace('Pemerintah Negara', 3, inplace=True)
data_train.KP.replace('Tanpa di Bayar', 4, inplace=True)
data_train.KP.replace('Tidak Pernah Bekerja', 4, inplace=True)
data_train.Gaji.replace('<=7jt', 0, inplace=True)
data_train.Gaji.replace('>7jt', 1, inplace=True)
#######################################################
data_test.JK.replace('Laki2', 0, inplace=True)
data_test.JK.replace('Perempuan', 1, inplace=True)
data_test.KP.replace('?', 1, inplace=True)
data_test.KP.replace('Wiraswasta', 1, inplace=True)
data_test.KP.replace('Pekerja Bebas Perusahaan', 2, inplace=True)
data_test.KP.replace('Pekerja Bebas Bukan Perusahan', 2, inplace=True)
data_test.KP.replace('Pemerintah Lokal', 3, inplace=True)
data_test.KP.replace('Pemerintah Provinsi', 3, inplace=True)
data_test.KP.replace('Pemerintah Negara', 3, inplace=True)
data_test.KP.replace('Tanpa di Bayar', 4, inplace=True)
data_test.KP.replace('Tidak Pernah Bekerja', 4, inplace=True)
data_train
```

```
y = data_train.Gaji
############################################
X = data_train.drop(columns=['Gaji'])
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```
from sklearn.impute import SimpleImputer
# Imputation
my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_test = pd.DataFrame(my_imputer.transform(X_test))

#imputed_X_train
#Imputation removed column names; put them back
imputed_X_train.columns = X_train.columns
imputed_X_test.columns = X_test.columns
#imputed_X_train.astype(int)
```

```
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, f1_score
from sklearn.neighbors import KNeighborsClassifier
```

```
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error ,explained_variance_score, mean_squared_error

parameters = {'max_depth':  list(range(6, 100, 10)),
              'max_leaf_nodes': list(range(50, 1001, 50))}

from sklearn.model_selection import GridSearchCV

gsearch = GridSearchCV(estimator=DecisionTreeClassifier(),
                       param_grid = parameters, 
                       scoring='f1',
                       n_jobs=5,cv=10,verbose=7)

gsearch.fit(imputed_X_train, y_train)
```

```
print(gsearch.best_params_.get('max_leaf_nodes'))
print(gsearch.best_params_.get('max_depth'))
```

```
def getMeanError():
    #forest_model = KNeighborsClassifier(n_neighbors=n_est)
    forest_model = DecisionTreeClassifier(max_leaf_nodes=gsearch.best_params_.get('max_leaf_nodes'),
                                          max_depth=gsearch.best_params_.get('max_depth'),
                                          random_state=1)
    forest_model.fit(imputed_X_train,y_train)
    preds = forest_model.predict(imputed_X_test)
    print(classification_report(y_test,preds))
    print(accuracy_score(y_test, preds))
    return f1_score(y_true=y_test, y_pred=preds)

getMeanError()
```

```
decisiontree = DecisionTreeClassifier()
decisiontree.fit(imputed_X_train, y_train)
y_pred = decisiontree.predict(data_test)
acc_decisiontree = round(decisiontree.score(imputed_X_train, y_train) * 100, 2)
print(acc_decisiontree)
test_out = pd.DataFrame({
    'Id': data_test.index, 
    'Gaji': y_pred
})
test_out.to_csv('submission5.csv', index=False)

print('Submission file has been exported.')
test_out.head()
```

```
gbk = GradientBoostingClassifier()
gbk.fit(imputed_X_train, y_train)
y_pred = gbk.predict(data_test)
acc_gbk = round(gbk.score(imputed_X_train, y_train) * 100, 2)
print(acc_gbk)
```

```
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(imputed_X_train, y_train)
Y_pred = knn.predict(data_test)
acc_knn = round(knn.score(imputed_X_train, y_train) * 100, 2)
acc_knn

```

```
svc = SVC()
svc.fit(imputed_X_train, y_train)
Y_pred = svc.predict(data_test)
acc_svc = round(svc.score(imputed_X_train, y_train) * 100, 2)
acc_svc
test_out = pd.DataFrame({
    'Id': data_test.index, 
    'Gaji': Y_pred
})
test_out.to_csv('submission4.csv', index=False)

print('Submission file has been exported.')
test_out.head()
```
